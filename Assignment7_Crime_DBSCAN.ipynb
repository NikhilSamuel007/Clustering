```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
crime = pd.read_csv("crime_data.csv")
```

```python
crime.head()
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

```python
array = crime.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
```

```python
X = stscaler.fit_transform(array)
X
```

```python
dbscan = DBSCAN(eps=1,min_samples=4)
```

```python
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
crime['clusterid']= dbscan.labels_
```

```python
crime
```

# -1 is noise samples

```python
crime.groupby('clusterid').mean()
```

```python
plt.figure(figsize=(15,9))
plt.scatter(crime['clusterid'],crime['UrbanPop'],c= dbscan.labels_)
```
